```python
import streamlit as st
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

st.title("S&P 500 Return and Seasonal Behavior Dashboard")

# Sidebar inputs
ticker = st.sidebar.text_input("Enter ticker (e.g., ^GSPC or SPY)", "^GSPC")
start_date = st.sidebar.date_input("Start date", pd.to_datetime("2020-01-01"))
end_date = st.sidebar.date_input("End date", pd.to_datetime("2020-12-31"))

# Download data
data = yf.download(ticker, start=start_date, end=end_date)
close = data['Close'].dropna()

# Calculate return
# Make sure you're working with a single column (Series)
close_series = data['Close'].dropna()

# Extract scalar values
start_price = float(close.iloc[0])
end_price = float(close.iloc[-1])

# Calculate return as a float
return_pct = (end_price - start_price) / start_price

# Display metric
st.metric(label="Total Return", value=f"{return_pct:.2%}")

# Now this will work
st.metric(label="Total Return", value=f"{return_pct:.2%}")


# Line chart
st.subheader("Daily Closing Prices")
st.line_chart(close)

# Upload PCA scatter plot
st.subheader("Monthly PCA Scatter Plot")
uploaded_file = st.file_uploader("Upload your PCA scatter plot image", type=["png", "jpg"])
if uploaded_file:
    st.image(uploaded_file, caption="Monthly PCA Clustering", use_column_width=True)
```
